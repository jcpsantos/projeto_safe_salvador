```python
#-> Importando bibliotecas necessárias 
import pdftables_api as pdf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
# Função para converter os arquivos pdf em csv
def converter(input_pdf, output_csv):
    """Converter tabelas PDF em tabelas CSV
      Argumentos:
          input_pdf: arquivo/tabela em PDF.
          output_csv: saída do aquivo em CSV.
      Retorna:
          Um arquivo/tabela no formato CSV"""
    c = pdf.Client('qxppgsjih80k')
    c.csv(input_pdf, output_csv)
```

```python
#Ativando a função de conversão
converter('03CAPITAL2014.pdf', '2014.csv')
converter('03CAPITAL2015.pdf', '2015.csv')
converter('03CAPITAL2016.pdf', '2016.csv')
converter('03_CAPITAL_2017_JAN_A_DEZ.pdf', '2017.csv')
converter('03_CAPITAL_2018_JAN_A_MAR.pdf', '2018.csv')
```

<h2 style="color: #f44336;">Limpeza dos dados</h2>

```python
# Excluindo as linhas com valores nulos
ds2014 = ds2014.dropna(axis=0)
ds2015 = ds2015.dropna(axis=0)
ds2016 = ds2016.dropna(axis=0)
ds2017 = ds2017.dropna(axis=0)
ds2018 = ds2018.dropna(axis=0)
```

```python
#Vamos verificar como ficou a tabela de 2014
ds2014.head()
```

```python
#Agora vamos melhora os nomes das colunas
new_labels = ['aisp', 'homicídio doloso', 'lesão corporal seguida de morte', 'latrocínio', 'tentativa de homicídio', 'estupro', 'roubo a ônibus', 'roubo de veículo', 'furto de veículo', 'uso/porte substância entorpecente ']
#Função para novos nomes nas colunas
def name_columns(df):
    """Incluir novas labels nas colunas
      Argumentos:
          df: Lista/base de dados.
      Retorna:
          Lista/base de dados."""
    df.columns = new_labels
```

```python
name_columns(ds2014)
name_columns(ds2015)
name_columns(ds2016)
name_columns(ds2017)
name_columns(ds2018)
```

```python
#Verificando se as alterações foram realizadas
ds2015.head()
```

```python
#Vamos verificar se tem açguma coluna com valores faltantes
print("Colunas com valores Faltantes: \n {0} \n".format(
    ds2016.columns[ds2016.isnull().any()].values))
```

```python
#Criando colunas do ano
# Criando vetor de ano para o dataframe
def ano_vetor(ano, ds):
    """Realizar uma lista com valores do ano
      Argumentos:
          ano: Valor do ano a ser incluído.
          ds: Lista/base de dados
      Retorna:
          Vetor/Lista com os valores do ano desejado."""
    x_ano = np.repeat(ano, ds.shape[0])
    return x_ano

ano_2014=ano_vetor(2014, ds2014)
print(ano_2014)

ano_2015=ano_vetor(2015, ds2015)
print(ano_2015)

ano_2016=ano_vetor(2016, ds2016)
print(ano_2016)

ano_2017=ano_vetor(2017, ds2017)
print(ano_2017)

ano_2018=ano_vetor(2018, ds2018)
print(ano_2018)
```

```python
#Adicionando os vetores a nova coluna de Ano
ds2014['ano'] = ano_2014
ds2015['ano'] = ano_2015
ds2016['ano'] = ano_2016
ds2017['ano'] = ano_2017
ds2018['ano'] = ano_2018
ds2014.head()
```

```python
#Combinndo os dataframes em um só
ds_total = ds2014.append(ds2015)
ds_total = ds_total.append(ds2016)
ds_total = ds_total.append(ds2017)
ds_total = ds_total.append(ds2018)

ds_total
```

```python
# converta as colunas  para int
def converter_int(df):
    """Converter valores do tipo string em int
      Argumentos:
          df: Lista/base de dados
      Retorna:
          Uma validação de que os valores foram transformados em inteiros."""
    df=df.str.extract('(\d+)').astype(int)
    return df.info()
```

```python
converter_int(ds_total['homicídio doloso'])
converter_int(ds_total['lesão corporal seguida de morte'])
converter_int(ds_total['latrocínio'])
converter_int(ds_total['tentativa de homicídio'])
converter_int(ds_total['estupro'])
converter_int(ds_total['roubo a ônibus'])
converter_int(ds_total['roubo de veículo'])
converter_int(ds_total['furto de veículo'])
```
