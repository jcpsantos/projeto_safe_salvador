**Vamos importar as bibliotecas necessárias converter os arquivos pdf em csv e carregar os arquivos csv e investigar os dados**

```python
#-> Importando bibliotecas necessárias 
import pdftables_api as pdf
import pandas as pd
```

```python
# Função para converter os arquivos pdf em csv
def converter(input_pdf, output_csv):
    """Converter tabelas PDF em tabelas CSV
      Argumentos:
          input_pdf: arquivo/tabela em PDF.
          output_csv: saída do aquivo em CSV.
      Retorna:
          Um arquivo/tabela no formato CSV"""
    c = pdf.Client('qxppgsjih80k')
    c.csv(input_pdf, output_csv)
```

```python
#Ativando a função de conversão
converter('03CAPITAL2014.pdf', '2014.csv')
converter('03CAPITAL2015.pdf', '2015.csv')
converter('03CAPITAL2016.pdf', '2016.csv')
converter('03_CAPITAL_2017_JAN_A_DEZ.pdf', '2017.csv')
converter('03_CAPITAL_2018_JAN_A_MAR.pdf', '2018.csv')
```

```python
#Carregando o arquivo de dados
ds2014 = pd.read_csv('2014.csv')
ds2015 = pd.read_csv('2015.csv')
ds2016 = pd.read_csv('2016.csv')
ds2017 = pd.read_csv('2017.csv')
ds2018 = pd.read_csv('2018.csv')

#Verificando os dados carregados de 2014
ds2014
```

```python
#numero de amostras  (linhas e colunas)
def amostras(ds, ano):
    """Mostra os valores de amostras e colunas do arquivo
      Argumentos:
          ds: Lista/base de dados.
          ano: Ano do arquivo
      Retorna:
          Um print informando os valores de amostras e colunas do arquivo."""
    print("O arquivo {0} possui {1} amostras e {2} colunas".format(
        ano, ds.shape[0], ds.shape[1] ))
amostras(ds2014, 2014)
amostras(ds2015, 2015)
amostras(ds2016, 2016)
amostras(ds2017, 2017)
amostras(ds2018, 2018)
```
